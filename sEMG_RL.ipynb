```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np

class ActorCritic(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(ActorCritic, self).__init__()
        # Shared layers
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)

        # Actor (policy) head
        self.actor = nn.Linear(hidden_size, output_size)
        self.softmax = nn.Softmax(dim=1)

        # Critic (value) head
        self.critic = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.dropout(x)

        # Actor output (action probabilities)
        action_probs = self.softmax(self.actor(x))

        # Critic output (state value)
        state_value = self.critic(x)

        return action_probs, state_value

# Hyperparameters
input_size = 64     # Number of sensor readings (EMG signals)
output_size = 4    # Number of action classes
hidden_size = 64
learning_rate = 0.001
epochs = 100
batch_size = 32

# Load your dataset (replace with actual data)
X_train = input_data_x.to_numpy()  # Input features (num_samples × 8)
y_train = input_data_y.to_numpy()  # Labels (num_samples × 1)

# Convert to PyTorch tensors
X_tensor = torch.tensor(X_train, dtype=torch.float32)
y_tensor = torch.tensor(y_train, dtype=torch.long)

# Create DataLoader
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Initialize network and optimizer
model = ActorCritic(input_size, hidden_size, output_size)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop for supervised pre-training
for epoch in range(epochs):
    total_loss = 0
    correct = 0
    total = 0

    for batch_X, batch_y in dataloader:
        # Get action probabilities and state value
        action_probs, state_value = model(batch_X)

        # Create categorical distribution from probabilities
        dist = torch.distributions.Categorical(action_probs)

        # Sample an action from the distribution
        actions = dist.sample()

        # Calculate rewards (+1 for correct, -1 for incorrect)
        rewards = (actions == batch_y).float() * 2 - 1

        # Calculate log probability of the taken actions
        log_probs = dist.log_prob(actions)

        # Calculate advantage (reward - state_value)
        advantage = rewards - state_value.detach()

        # Calculate actor and critic losses
        actor_loss = -torch.mean(log_probs * advantage)
        critic_loss = nn.MSELoss()(state_value, rewards)

        # Total loss
        loss = actor_loss + critic_loss

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update metrics
        total_loss += loss.item()
        correct += (actions == batch_y).sum().item()
        total += batch_y.size(0)

    # Calculate epoch statistics
    avg_loss = total_loss / len(dataloader)
    accuracy = correct / total

    print(f'Epoch [{epoch+1}/{epochs}], '
          f'Loss: {avg_loss:.4f}, '
          f'Accuracy: {accuracy:.4f}')

# Save the pre-trained model
torch.save(model.state_dict(), "pretrained_actor_critic.pth")
```

```python
import torch
import numpy as np
import pandas as pd
from torch.utils.data import DataLoader, TensorDataset

# Load the pre-trained model
model = ActorCritic(input_size, hidden_size, output_size)
model.load_state_dict(torch.load("pretrained_actor_critic.pth"))
model.train()  # Set the model to training mode

# Load the new dataset (values and labels separated)
# Replace "new_emg_values.csv" and "new_emg_labels.csv" with your actual files
#new_values = pd.read_csv("new_emg_values.csv")  # EMG signal features
#new_labels = pd.read_csv("new_emg_labels.csv")  # Corresponding labels

# Convert to numpy arrays
X_new = user_data_x.values  # EMG signal features (num_samples × input_size)
y_new = user_data_y.values.flatten()  # Labels (num_samples × 1)

# Convert to PyTorch tensors
X_new_tensor = torch.tensor(X_new, dtype=torch.float32)
y_new_tensor = torch.tensor(y_new, dtype=torch.long)

# Create DataLoader for the new dataset
new_dataset = TensorDataset(X_new_tensor, y_new_tensor)
new_dataloader = DataLoader(new_dataset, batch_size=batch_size, shuffle=True)

# Function to compute rewards based on correctness
def compute_rewards(predictions, actual_labels):
    rewards = (predictions == actual_labels).float() * 2 - 1  # +1 for correct, -1 for incorrect
    return rewards

# Online adaptation loop
online_epochs = 50
for epoch in range(online_epochs):
    total_loss = 0
    correct = 0
    total = 0

    # Iterate over the new dataset
    for batch_X_new, batch_y_new in new_dataloader:
        action_probs, state_value = model(batch_X_new)

        # Create categorical distribution from probabilities
        dist = torch.distributions.Categorical(action_probs)

        # Sample an action from the distribution
        actions = dist.sample()

        # Compute rewards based on correctness
        rewards = compute_rewards(actions, batch_y_new)

        # Calculate log probability of the taken actions
        log_probs = dist.log_prob(actions)

        # Calculate advantage (reward - state_value)
        advantage = rewards - state_value.detach()

        # Calculate actor and critic losses
        actor_loss = -torch.mean(log_probs * advantage)
        critic_loss = nn.MSELoss()(state_value, rewards)

        # Total loss
        loss = actor_loss + critic_loss

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update metrics
        total_loss += loss.item()
        correct += (rewards == 1).sum().item()
        total += len(batch_X_new)

    # Calculate epoch statistics
    avg_loss = total_loss / len(new_dataloader)
    accuracy = correct / total

    print(f'Online Epoch [{epoch+1}/{online_epochs}], '
          f'Loss: {avg_loss:.4f}, '
          f'Accuracy: {accuracy:.4f}')

# Save the fine-tuned model
torch.save(model.state_dict(), "fine_tuned_actor_critic.pth")
```
